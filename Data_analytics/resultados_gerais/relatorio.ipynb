```python
import matplotlib 
size_256x = []
size_6x =[]
size_256y = []
size_6y =[]
fig, ax = plt.subplots(figsize=(10,5),dpi=72)
ax.grid()
for i in array:
  if (i[2] == 256):
    size_256x.append(i[0])
    size_256y.append(i[1])
    plt.annotate(f'Min: {i[3]}\nMax: {i[4]}', (i[0], i[1]), xytext=(5, 5), textcoords='offset points')
  if (i[2] == 6):
    size_6x.append(i[0])
    size_6y.append(i[1]) 
    plt.annotate(f'Min: {i[3]}\nMax: {i[4]}', (i[0], i[1]), xytext=(5, -20), textcoords='offset points')

  ax.scatter(i[0],i[1], label='Dispositivo-{}ms'.format(i[:2]))



ax.plot(size_6x, size_6y, label="6 bytes")
ax.plot(size_256x, size_256y, label="256 bytes")
ax.legend()

  # i+=1
```
